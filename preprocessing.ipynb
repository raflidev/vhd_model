```python
import os
import shutil
import random

# Konfigurasi path
base_dir = 'dataset'
train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')

# List kelas yang akan diproses
classes = ['AS', 'MR', 'MS', 'MVP', 'N']

# Persentase split
split_ratio = 0.1

# Membuat folder test jika belum ada
if not os.path.exists(test_dir):
    os.makedirs(test_dir)

for class_name in classes:
    # Path sumber (train) dan tujuan (test) per kelas
    src_path = os.path.join(train_dir, class_name)
    dest_path = os.path.join(test_dir, class_name)
    
    # Buat folder destinasi per kelas jika belum ada
    if not os.path.exists(dest_path):
        os.makedirs(dest_path)
    
    # Ambil semua file di folder source
    if os.path.exists(src_path):
        files = [f for f in os.listdir(src_path) if os.path.isfile(os.path.join(src_path, f))]
        
        # Hitung jumlah file yang akan dipindahkan
        num_to_move = int(len(files) * split_ratio)
        
        # Pilih file secara acak
        files_to_move = random.sample(files, num_to_move)
        
        print(f"Processing Class: {class_name}")
        print(f"Total files: {len(files)}, Moving: {num_to_move}")
        
        # Pindahkan file
        for file_name in files_to_move:
            shutil.move(os.path.join(src_path, file_name), os.path.join(dest_path, file_name))
            
        print(f"Finished moving {class_name}")
    else:
        print(f"Directory not found: {src_path}")
```
